```python
random.seed("Xingdong")
primes = []
composites = []
lastDigits = [1,3,7,9]
start, end = 2**200, 2**201
for _ in range(100):
    for _ in range(10):
        number = random.randint(start, end)
        for lastDigit in lastDigits:
            newNumber = number * 10 + lastDigit
            if sympy.isprime(newNumber):
                primes.append(newNumber)
            else:
                composites.append(newNumber)
        start *= 2
        end *= 2
print(len(primes), len(composites))    
```

```python
aArr = []
for i in range(2, 10000):
    if sympy.isprime(i):
        aArr.append(i)
print(len(aArr))
```

```python
results = []
for c in composites:
    for a in aArr:
        if gcd(c, a) != 1:
            results.append(True)
            print(str(c) + " is a composite")
            break
        if pow(a, c-1, c) == 1:
            results.append(True)
            print(str(c) + " is a composite")
            break
    else:
        results.append(False)
        print(str(c) + " is a prime")
print(results.count(True)/len(results)*100)
```

```python
def gcd(a, b):
    if a < b:
        return gcd(b, a)
    if a % b == 0:
        return b
    return gcd(b, a%b)
```

```python
def cal_s_d(n):
    n -= 1
    s = 0
    while n & 1 == 0:
        s += 1
        n >>= 1
    return s, n
print(cal_s_d(13))
print(cal_s_d(17))
```

```python
results = []
for c in primes:
    s, d = cal_s_d(c)
    for a in aArr:
        if pow(a, d, c) == 1:
            continue
        for r in range(s):
            if pow(a, 2**r*d, c) == c-1:
                break
        else:           
            results.append(True)
            print(str(c) + " is a composite")
            break
    else:
        results.append(False)
        print(str(c) + " is a prime")
print(results.count(True)/len(results)*100)
```
